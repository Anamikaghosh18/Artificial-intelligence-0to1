```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv(r'C:\Desktop\Anamika\ML0to1\Datasets\heart.csv')
```

```python
df
```

```python
X = df.iloc[:, 0:13]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

lr = LogisticRegression()
dt = DecisionTreeClassifier()

model1 = lr.fit(X_train, y_train)
model2 = dt.fit(X_train, y_train)
```

```python
lr_pred = model1.predict(X_test)
dt_pred = model2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix, precision_score, recall_score, f1_score

print(f"Accuracy score for Logistic regression: {accuracy_score(lr_pred, y_test)}")
print(f"Accuracy score for Decision tree: {accuracy_score(dt_pred, y_test)}")
```

```python
print(f"confusion_matrix for logistic : {confusion_matrix(lr_pred, y_test)}")
print(f"Confusion matrix for decision trees : {confusion_matrix(dt_pred, y_test)}")
```

```python
print(f"precision for logistic : {precision_score(lr_pred, y_test)}")
print(f"precision for decision trees : {precision_score(dt_pred, y_test)}")
```

```python
print(f"recall score for logistic : {recall_score(lr_pred, y_test)}")
print(f"Recall for decision trees : {recall_score(dt_pred, y_test)}")
```

```python
print(f"f1 score for logistic : {f1_score(lr_pred, y_test)}")
print(f"f1 score for decision trees : {f1_score(dt_pred, y_test)}")
```
