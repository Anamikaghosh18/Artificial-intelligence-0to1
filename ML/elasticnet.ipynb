```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import ElasticNet, LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y = True)
```

```python
X_train , X_test ,y_train,  y_test = train_test_split(X, y , test_size=0.2 , random_state=2)
```

```python
reg = LinearRegression()
regmodel = reg.fit(X_train, y_train)
y_pred = regmodel.predict(X_test)
print(f"R2 score: {r2_score(y_test, y_pred)}")
```

```python
ridge = Ridge(alpha = 0.1)
ridgemodel = ridge.fit(X_train, y_train)

ridge_pred = ridgemodel.predict(X_test)

print(f'R2score : {r2_score(y_test, ridge_pred)}')
```

```python
lasso = Lasso(alpha = 0.01)
lassomodel = lasso.fit(X_train, y_train)

lasso_pred = lassomodel.predict(X_test)

print(f'R2score : {r2_score(y_test, lasso_pred)}')
```

```python
elasticnet = ElasticNet(alpha = 0.005, l1_ratio = 0.9)
elastic_model = lasso.fit(X_train, y_train)

elastic_pred = elastic_model.predict(X_test)

print(f'R2score : {r2_score(y_test, elastic_pred)}')
```
