```python
import numpy as np 
import pandas as pd 
from  sklearn.datasets import load_diabetes
data = load_diabetes()
```

```python
print(data.DESCR)
```

```python
x = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(x,y, random_state=2, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression 

lr = LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(f"R2score : {r2_score(y_test, pred)}")
print(f"Mean squared error: {np.sqrt(mean_squared_error(y_test, pred))}")
```

```python
from sklearn.linear_model import Ridge 

R = Ridge(alpha = 1.0)

```

```python
ridge_model = R.fit(X_train, y_train)
```

```python
ridge_pred = ridge_model.predict(X_test)
```

```python
print(f'Mean squared error : {np.sqrt(mean_squared_error(y_test, ridge_pred))}')
print(f'R2 score : {r2_score(y_test, ridge_pred)}')
```
