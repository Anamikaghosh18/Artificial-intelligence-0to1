```python
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import StandardScaler , PolynomialFeatures

from sklearn.metrics import r2_score 

from sklearn.pipeline import Pipeline
  
```

```python
X = 6 * np.random.rand(200, 1) - 3 

y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y, random_state=2, test_size= 0.2)
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(X , y)
```

```python
y_pred = lr.predict(X_test)

```

```python
print(f"R2_score:{r2_score(y_test,y_pred)}")
```

```python
plt.plot(X_train, lr.predict(X_train), color = 'r')
plt.scatter(X_train, y_train)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
# applying polynomial linear regression 

# polynomial feature s always applied to input features cant be applied on  output features 

poly = PolynomialFeatures(degree = 2)

X_train_transformed  = poly.fit_transform(X_train)
X_test_transformed = poly.fit_transform(X_test)
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(X_train_transformed, y_train)
```

```python
y_predited = lr.predict(X_test_transformed)
```

```python
print(f"R2 score: {r2_score(y_test, y_predited)}")
```
