## Simple Linear Regression 
- In SLR we try to draw a best fit line on sort of linear data , which does less errors.


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Desktop\Anamika\ML0to1\Datasets\placement.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x = 'cgpa', y = 'package',data = df)
plt.xlabel('cgpa')
plt.ylabel('package')
```

 - Before applying any algo we seaparate the input and output columns 

```python
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 2)

lr = LinearRegression()
model = lr.fit(X_train, y_train)
```

```python
predict = lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
sns.scatterplot(x = 'cgpa', y = 'package', data = df)
plt.plot(X_test,lr.predict(X_test), color = 'red')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')

```

```python
m = lr.coef_ # slope 
```

```python
c = lr.intercept_
```

```python
y = m * 4.1 + c
print(y)
```

## saving and using the model 

```python
import joblib

joblib.dump(model, 'package_predictor')
```
